```python
#Firing Up Turicreate
```

```python
import turicreate
```

```python
sf = turicreate.SFrame('home_data.sframe')
```

```python
sf.show()
```

```python
sf
```

```python
x = sf['sqft_living']
```

```python
y = sf['price']
```

```python
turicreate.visualization.scatter(x,y)
```

```python
#creating regression model
```

```python
train_data, test_data = sf.random_split(.8, seed=0)
```

```python
sqft_model = turicreate.regression.create(train_data, target='price', features=['sqft_living'])
```

```python
#Evaluating the Model
```

```python
print(test_data)
```

```python
print(test_data['price'].mean())
```

```python
print(sqft_model.evaluate(test_data))
```

```python
print(sqft_model)
```

```python
#Showing Prediction
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.plot(test_data['sqft_living'], test_data['price'], '.', test_data['sqft_living'], sqft_model.predict(test_data), '-')
```
